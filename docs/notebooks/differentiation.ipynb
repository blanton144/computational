```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def diff_central(func=None, x=None, dx=None):
    return((func(x + 0.5 * dx) - func(x - 0.5 * dx)) / dx)
```

```python
num = 1000
x = 2. * np.pi * (np.arange(num, dtype=np.float64) + 0.5) / np.float64(num)
f = np.sin(x) 
dfdx = diff_central(func=np.sin, x=x, dx=1.e-5)
```

```python
plt.plot(x, f)
```

```python
plt.plot(x, dfdx)
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, dfdx - np.cos(x))
plt.ylim((-1.e-10, 1.e-10))
```

```python
def diff_correct(func=None, x=None, dx=None):
    return((4. * diff_central(func=func, x=x, dx=dx * 0.5) -
            diff_central(func=func, x=x, dx=dx)) / 3.)
```

```python
num = 1000
x = 2. * np.pi * (np.arange(num, dtype=np.float64) + 0.5) / np.float64(num)
f = np.sin(x) 
dfdx = diff_correct(func=np.sin, x=x, dx=3.e-3)
```

```python
plt.plot(x, dfdx - np.cos(x))
plt.ylim((-1.e-13, 1.e-13))
```
